```python
from Neural_Networks import * 
import cv2
import numpy as np
import matplotlib.pyplot as plt 

model = Model.load('fashion_mnist.model')
```

```python
# Now to use or model to predict
fashion_mnist_labels ={0: 'T-shirt/top',
1: 'Trouser',
2: 'Pullover',
3: 'Dress', 
4: 'Coat' ,
5: 'Sandal', 
6: 'Shirt', 
7: 'Sneaker', 
8: 'Bag',
9: 'Ankle boot' }

def predict(data):
    image_data = cv2.imread(data, cv2.IMREAD_GRAYSCALE)

    plt.imshow(image_data, cmap='gray')
    plt.show()

    image_data = cv2.resize(image_data, (28,28))
    image_data = 255 - image_data # as the colours are inverted in training data and prediction example
    image_data = (image_data.reshape(1,-1).astype(np.float32) - 127.5) / 127.5

    confidences = model.predict(image_data)
    predictions = model.output_layer_activation.predictions(confidences)

    for prediction in predictions:
        print(f'prediction: {fashion_mnist_labels[prediction]}')
```

```python

predict('tshirt.png')
```
